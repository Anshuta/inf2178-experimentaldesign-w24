## Mixed Effect Anova Analysis for research question 1:
## How does the Normalized Whole Brain Volume (nWBV) vary across different visits for each group (Demented, Nondemented and converted), and is there an interaction effect between visits and group status on nWBV changes?

```python
# Create the point plot to see nWBV between each group and visit
sns.set()
sns.pointplot(data=df_brain, x='Visit', y='nWBV', hue='Group', dodge=True, markers=['o', 's','D'],
	      capsize=.1, errwidth=1, palette='colorblind')
```

```python
df_brain.groupby(['Visit', 'Group'])['nWBV'].agg(['mean', 'std']).round(2)
```

```python
!pip install pingouin
import pingouin as pg
# Compute the two-way mixed-design ANOVA
pg.mixed_anova(dv='nWBV', within='Visit', between='Group', subject='Subject ID', data=df_brain)

```

```python
# Perform the post hoc tests
posthocs = pg.pairwise_ttests(dv='nWBV', within='Visit', between='Group',
                              subject='Subject ID', data=df_brain)
pg.print_table(posthocs)
```

```python
#Checking assumptions
#mauchly's test of spericity
pg.sphericity(data=df_brain, dv='nWBV', subject='Subject ID', within='Visit')[-1]
```

```python
#Test of normality
pg.normality(data=df_brain, dv='nWBV', group='Visit')
```

```python
#Homogeneity of variance test(Levene)
pg.homoscedasticity(data=df_brain, dv='nWBV', group='Visit')
```

## Power plots analysis for t test

```python
from math import sqrt
from statsmodels.stats.power import TTestIndPower

# List parameters
effect_size = 0.7
alpha = 0.05
power = 0.91
# perform power analysis to find sample size
power_analysis = TTestIndPower()
sample_size = power_analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=1)
print('Sample size/Number needed in each group: {:.3f}'.format(sample_size))
```

```python
#power curves
# power analysis parameters
effect_sizes = np.array([effect_size])
sample_sizes = np.array(range(5, 100))
# plot power curves
obj = TTestIndPower()
obj.plot_power(dep_var='nobs', nobs=sample_sizes,
			effect_size=effect_sizes)

plt.show()
```
